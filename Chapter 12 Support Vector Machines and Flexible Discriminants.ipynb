# Ex. 12.1

Let's define $f(x)=x_i^T\beta+\beta_0.$ Then 12.8 is:

\begin{align*}
\min_{\beta,\beta_{0},\xi_{i}} & \quad\frac{1}{2}\left\Vert \beta\right\Vert ^{2}+C\sum_{i=1}^{N}\xi_{i}\\
\mbox{subject to} & \quad\xi_{i}\ge0,\\
 & \quad1-y_{i}f(x_{i})\le\xi_{i}\quad\forall i
\end{align*}

Setting $\lambda=1/C$, 12.25 becomes:

\begin{align*}
\min_{\beta,\beta_{0}} & \quad\frac{1}{2}\left\Vert \beta\right\Vert ^{2}+C\sum_{i=1}^{N}\left(1-y_{i}f(x_{i})\right)^{+}
\end{align*}

It is easily observed that the constraints in 12.8 is equivalent to:
$$\xi_{i}\ge\left(1-y_{i}f(x_{i})\right)^{+}\quad\forall i.$$

Next we show that 12.25 is the partial optimal solution after minimizing w.r.t. $\xi_i$:

Let $\beta,\beta_0$ be arbitrarily chosen and fixed. Let's try to minimize the objective by choosing the optimal $\xi_i$. 

With $\beta,\beta_0$ fixed, the objective is only a function of $\xi_i$ and we simply select the smallest possible $\xi_i$ for each $i$: 
$$\xi_{i} = \left(1-y_{i}f(x_{i})\right)^{+}.$$

We update the objective by plugging in the optimal value of $C\sum_{i=1}^N\xi_i$ to obtain the desired result.

# Ex. 12. 2

When penalizing without the intercept, the particular kernel to make 12.29 the same as 12.25 is the linear kernel:
$$ K(x,x') = x^Tx'. $$

We are going to show that, under this kernel, the norm of the function (without the intercept $\beta_0$), $J(f-\beta_0)$ is exactly the $L_2$ norm of $\beta$.
As a result the two minimization problem become the same.

Let's first derive the norm of $f$ in $\mathcal{H}_K$. With the conclusion in Ex. 5.15, we have: 

\begin{align*}
J(f-\beta_0) & =\langle f-\beta_0,f-\beta_0\rangle_{\mathcal{H}_{K}}\\
 & =\sum_{i=1}^{N}\sum_{j=1}^{N}\alpha_{i}\alpha_{j}K(x_{i},x_{j})
\end{align*}

With the linear kernel plugged in, we have:

\begin{align*}
J(f-\beta_0) & =\sum_{i=1}^{N}\sum_{j=1}^{N}\alpha_{i}\alpha_{j}K(x_{i},x_{j})\\
 & =\sum_{i=1}^{N}\sum_{j=1}^{N}\alpha_{i}x_{i}^{T}x_{j}\alpha_{j}\\
 & =\sum_{i=1}^{N}\sum_{j=1}^{N}\sum_{k=1}^{p}\alpha_{i}x_{ik}x_{jk}\alpha_{j}
\end{align*}

Next we look at the function $f$ itself:

\begin{align*}
f(x) & =\sum_{i=0}^{N}\alpha_{i}K(x,x_{i})+\beta_{0}\\
 & =\sum_{i=0}^{N}\alpha_{i}x_{i}^{T}x+\beta_{0}\\
 & =\beta^{T}x+\beta_{0},
\end{align*}

where
$\beta=\sum_{i=1}^{N}\alpha_{i}x_{i}$ and $\beta_{k}=\sum_{i=1}^{N}\alpha_{i}x_{ik}.$

As a result:

\begin{align*}
\left\Vert \beta\right\Vert _{L_{2}}^{2} & =\left\langle \beta,\beta\right\rangle _{L_{2}}\\
 & =\sum_{k=1}^{p}\beta_{k}^{2}\\
 & =\sum_{k=1}^{p}\left(\sum_{i=1}^{N}\alpha_{i}x_{ik}\right)^{2}\\
 & =\sum_{k=1}^{p}\sum_{i=1}^{N}\alpha_{i}x_{ik}\sum_{j=1}^{N}\alpha_{j}x_{jk}\\
 & =\sum_{i=1}^{N}\sum_{j=1}^{N}\sum_{k=1}^{p}\alpha_{i}x_{ik}x_{jk}\alpha_{j}.
\end{align*}

This proves that when we choose the kernel: $K(x,x') = x^Tx',$
$$\left\Vert \beta\right\Vert_{L_{2}}^{2}=\left\Vert f\right\Vert_{\mathcal{H}_K}^{2}.$$

This will make these two optimization problem exactly the same.
